```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

L = 0.01 # Henry
C = 0.01 # Farad

def impedanceParallelRLC(omega, R, L, C):
    return 1 / (1/R - 1j * omega * L - 1/(1j * omega * C))

omega = np.arange(1, 10000.0, 10)

# log x and y axis
plt.loglog(omega, np.abs(impedanceParallelRLC(omega, 0.02, L, C)))
plt.loglog(omega, np.abs(impedanceParallelRLC(omega, 0.20, L, C)))
plt.loglog(omega, np.abs(impedanceParallelRLC(omega, 2.00, L, C)))
plt.title('Parallel RLC')
plt.show()
```

```python
def impedanceSeriesRLC(omega, R, L, C):
    return R - 1j * (omega * L - 1.0 / (omega * C))

def impedanceSeriesRLC2(omega, R, L, C):
    omega_0 = np.sqrt(1.0 / (L*C))
    Q = R / (2.0 * L)
    return ((omega**2 - omega_0**2) + 1j * 2.0 * Q * omega * omega_0)

plt.loglog(omega, np.abs(impedanceSeriesRLC(omega, 0.02, L, C)))
plt.loglog(omega, np.abs(impedanceSeriesRLC2(omega, 0.02, L, C)))
plt.show()
```

```python
def highpass(omega, omega_0):
    return omega / (omega - omega_0)

plt.loglog(omega, np.abs(highpass(omega, 1000.0)))
plt.loglog(omega, np.abs(highpass(omega, 1000.0)))
plt.show()
```
